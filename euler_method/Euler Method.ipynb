```python
# def func(x):
#     '''
#     inputs:
#         x: the input values
    
#     output:
#     f(x): the square of the inputs'''
#     return np.sin(x)
```

```python
def f(x):
    '''
    inputs:
        x: the input values
    
    output:
        df(x)/dx: the derivative of the function x^2'''
    return 2.0 * x * (1 - (x/30))
#     return np.cos(x)
```

```python
def euler(x0, y, h, x): 
    temp = -0
  
    # Iterating till the point at which we 
    # need approximation 
    while x0 < x: 
        temp = y 
        y = y + h * f(y) 
        x0 = x0 + h 
    
    
    return y
```

```python
def plot_function(x, res):
    '''inputs:
        x: the x values of the function [x_1, x_2, ..., x_n]
        y: the function values at point x [f(x_1), f(x_2), ..., f(x_n)]'''
    fig = plt.figure()
    ax = plt.axes()
    plt.grid(b=True)
    plt.plot(x,res, label='euler')
    plt.legend()
    plt.show()
```

```python
x_n = 100
x = np.linspace(1, 100, x_n)
```

```python
# Initial Values 
# x0 = 0
y0 = 10
h = 1 #step value h = (x(n) - x(0))/n(smaller better approximation)
res = []
res.append(y0)

# Result value with euler approximation
for i in range(x_n-1):
    temp = euler(i, y0, h, i+1)
    res.append(temp)
    y0 = temp
```

```python
plot_function(x, res) # Too much low h, bad approximation
```

```python
# Initial Values 
# x0 = 0
y0 = 10
h = 0.01 #step value h = (x(n) - x(0))/n(smaller better approximation)
res = []
res.append(y0)

# Result value with euler approximation
for i in range(x_n-1):
    temp = euler(i, y0, h, i+1)
    res.append(temp)
    y0 = temp
```

```python
plot_function(x, res)
```
